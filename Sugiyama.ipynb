```python
import grandalf as grand
from grandalf.layouts import SugiyamaLayout
import networkx as nx
import grandalf
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc('figure', figsize=[140,80])
```

```python
class defaultview(object):
    w, h = 2, 2
    
def layout(g, routing_count=1):
    if isinstance(g, nx.Graph):
        nxg = g
        gg = grandalf.utils.convert_nextworkx_graph_to_grandalf(g)
    else:
        nxg = grandalf.utils.convert_grandalf_graph_to_networkx_graph(g)
        gg = g
    
    for v in gg.V(): v.view = defaultview()

    sug = SugiyamaLayout(gg.C[0])
    sug.init_all() 
    sug.draw(routing_count) 

    positions = {v.data: (v.view.xy[0], v.view.xy[1]) for v in gg.C[0].sV} 
    nx.draw(nxg, pos=positions, with_labels=False)
    plt.show()
```

```python
g = nx.connected_watts_strogatz_graph(500,3,p=.320)  
layout(g,2)
```

```python
from grandalf.graphs import Vertex, Edge, Graph
from grandalf.layouts import VertexViewer
def create_scenario():
    '''
    Create something as:
    v4      v0
     \     / |
      \   v1 |
       \ /   |
       v5   /
        \  /
        v2
        |
        v3
    '''
    E = []
    data_to_vertex = {}

    vertices = []
    for i in range(6):
        data = 'v%s' % (i,)
        v = Vertex(data)
        data_to_vertex[data] = v
        v.view = VertexViewer(100, 50)
        vertices.append(v)

    edge = Edge(vertices[0], vertices[1])
    edge.view = EdgeViewer()
    E.append(edge)

    edge = Edge(vertices[0], vertices[2])
    edge.view = EdgeViewer()
    E.append(edge)

    edge = Edge(vertices[1], vertices[5])
    edge.view = EdgeViewer()
    E.append(edge)

    edge = Edge(vertices[2], vertices[3])
    edge.view = EdgeViewer()
    E.append(edge)

    edge = Edge(vertices[4], vertices[5])
    edge.view = EdgeViewer()
    E.append(edge)

    edge = Edge(vertices[5], vertices[2])
    edge.view = EdgeViewer()
    E.append(edge)

    G = Graph(vertices, E)
    assert len(G.C) == 1
    

    # not needed anymore...
    #r = filter(lambda x: len(x.e_in()) == 0, gr.sV)
    #if len(r) == 0:
    #    r = [gr.sV[0]]
    return G, data_to_vertex
```

```python
g = create_scenario()[0]
layout(g)
```
